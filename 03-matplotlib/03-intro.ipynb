```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x ** 2
y
```

## Plot Appearence

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

# color = RGB Hex Code or basic color names
# linewidth or lw = int or float
# alpha = float
# linestyle or ls = string
ax.plot(
    x, y,
    color='#FF8C00', linewidth=3, linestyle=':',
    marker='*', markersize=20, markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='purple', lw=2, ls='--')
ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```

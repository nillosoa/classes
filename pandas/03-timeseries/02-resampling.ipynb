```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('walmart_stock.csv')
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date']) # Same as df['Date'] = df['Date'].apply(pd.to_datetime)
df.info()
```

```python
df.set_index('Date', inplace=True)
df
```

```python
# Or simply
df = pd.read_csv('walmart_stock.csv', index_col='Date', parse_dates=True)
df
```

```python
df.index
```

```python
df.resample(rule='A').mean()
```

```python
df.resample(rule='BQ').mean()
```

```python
df.resample(rule='A').max()
```

```python
df.resample(rule='A').apply(lambda x: x[0]) # The first day
```

```python
df['Close'].resample('A').mean().plot(kind='bar')
```

```python
df['Close'].resample('M').mean().plot(kind='bar', figsize=(16, 6))
```
